The .zip file containing the CSV files for our project contain three CSVs in total. In order to combine them, we have to load them individually.

The processPiece function below has the simple goal of:

Renaming the columns to titles that are more readable
Grabbing only the top 10% of emissions in the nation for each CSV (based on total emissions)

```python
def processPiece(frame):
  neededColumns = ['city','state','total emissions','emissions uom','pollutant code','pollutant desc','pollutant type(s)','company name','site name','site longitude','site latitude']
  piece = frame[neededColumns]
  columnList = list(piece.columns)
  piece.columns = [c.replace(" ","")for c in columnList]
  piece = piece.sort_values(by=['totalemissions'],ascending=[False])
  valueCounts = piece['state'].value_counts().rename_axis('unique_values').reset_index(name="counts")
  valueCounts.columns = ["uniquevalues","counts"]
  valueCounts['counts'] = [int(v*0.1) for v in valueCounts['counts']]

  piece['state'] = piece['state'].fillna('None')
  spliceFrame=pd.DataFrame(columns=[n.replace(" ","") for n in neededColumns])


  for row in valueCounts.itertuples():
    spliceFrame = pd.concat([spliceFrame,piece[piece['state'].str.contains(row.uniquevalues)].head(int(row.counts))],axis=0)

  spliceFrame.columns = ['city','state','totalemissions','emissionsuom','pollutantcode','pollutantdesc','pollutanttypes','companyname','sitename','sitelongitude','sitelatitude']
  return spliceFrame
```

```python
def processSecondPiece(frame):
  neededColumns = ['city','state','total_emissions','emissions_uom','pollutant_code','pollutant_desc','pollutant_type','company_name','site_name','site_longitude','site_latitude']
  piece = frame[neededColumns]
  columnList = list(piece.columns)
  piece.columns = [c.replace(" ","")for c in columnList]
  piece = piece.sort_values(by=['total_emissions'],ascending=[False])
  valueCounts = piece['state'].value_counts().rename_axis('unique_values').reset_index(name="counts")
  valueCounts.columns = ["uniquevalues","counts"]
  valueCounts['counts'] = [int(v*0.1) for v in valueCounts['counts']]

  piece['state'] = piece['state'].fillna('None')
  spliceFrame=pd.DataFrame(columns=[n.replace(" ","") for n in neededColumns])


  for row in valueCounts.itertuples():
    spliceFrame = pd.concat([spliceFrame,piece[piece['state'].str.contains(row.uniquevalues)].head(int(row.counts))],axis=0)

  spliceFrame.columns = ['city','state','totalemissions','emissionsuom','pollutantcode','pollutantdesc','pollutanttypes','companyname','sitename','sitelongitude','sitelatitude']
  return spliceFrame
```

Looks daunting, but is actually fairly simple:

- Grabbing all company names, coordinates, emissions, and storing them as a JSON file, mapped by city and state.
- Filtering them by site names. All site names are unique and do not have double entries.

```python
def siteRestructuring(finalFrame):
  finalDictionary = {}
  for row in finalFrame.itertuples():
    emissionFactor = 2000 if row.emissionsuom == 'TON' else 1
    potentialKey = str(row.sitename)
    if potentialKey in finalDictionary.keys():
      dictionaryEntry = finalDictionary[potentialKey]
      dictionaryEntry['totalemissions']+=(row.totalemissions*emissionFactor)
      if(row.totalemissions>dictionaryEntry['largestemission']):
        dictionaryEntry['pollutanttype']=row.pollutanttypes
        dictionaryEntry['pollutantdesc']=row.pollutantdesc
        dictionaryEntry['largestemission']=row.totalemissions
    else:
      city = row.city
      state = row.state
      latitude = row.sitelatitude
      longitude = row.sitelongitude
      pollutantdesc=row.pollutantdesc
      pollutanttype=row.pollutanttypes
      totalemissions = row.totalemissions*emissionFactor
      finalDictionary[potentialKey] ={
          "city":city,
          "state":state,
          "latitude":latitude,
          "longitude":longitude,
          "pollutantdesc":pollutantdesc,
          "largestemission":totalemissions,
          "pollutanttype":pollutanttype,
          "totalemissions":totalemissions
      }
  return finalDictionary
```

```python
siteDictionary = siteRestructuring(finalPiece)
```
